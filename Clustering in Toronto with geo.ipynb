### Get Data

```python


url= 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

response=requests.get(url)

html = response.text

soup = BeautifulSoup(html,'lxml')


```

### Clean

```python
div = soup.find('div',class_ = "mw-parser-output")
table = div.table.text

tables = table.split(',')
tables = table.split('\n')


while '' in tables:
    tables.remove('')

print (tables)

len(tables)

```

### Build Dataframe

```python

from pandas.core.frame import DataFrame
tables1, tables2, tables3 = tables[3::3], tables[4::3], tables[5::3]

print(tables1)
print(tables2)
print(tables3)

# define the dataframe columns
column_names = ['PostalCode','Borough', 'Neighborhood'] 

# instantiate the dataframe
data = {"PostalCode":tables1,"Borough":tables2,"Neighborhood":tables3}


df = DataFrame(data,columns=column_names)

print(df)
```

### Delete not assigned

```python
df =df[~df['Borough'].str.contains('Not assigned')]

print(df)
```

### Combine Neighborhood

```python
df_1= df.groupby(by=['PostalCode','Borough'], as_index=False, sort=False).agg(lambda x: ','.join(x))

                                                

df_1
```
